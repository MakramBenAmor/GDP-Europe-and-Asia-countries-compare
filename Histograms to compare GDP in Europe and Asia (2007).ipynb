# 1St problem: compare Europe and Asia GDP (2007)

```python

import pandas as pd 
from matplotlib import pyplot as plt 
data= pd.read_csv("countries.csv")
data
```

```python
set(data.continent)
```

```python
data_2007=data[data.year== 2007]
```

```python
asia_data_2007=data_2007[data_2007.continent=="Asia"]
Europe_data_2007=data_2007[data_2007.continent=="Europe"]
```

```python
len(set(asia_data_2007.country))
```

```python
len(set(Europe_data_2007.country))
```

```python
print("Mean GDP per Capita in Asia: ")
print(asia_data_2007.gdpPerCapita.mean())

print("Mean GDP per Capita in Europe: ")
print(Europe_data_2007.gdpPerCapita.mean())

print("Median GDP per Capita in Asia: ")
print(asia_data_2007.gdpPerCapita.median())

print("Median GDP per Capita in Europe: ")
print(Europe_data_2007.gdpPerCapita.median())
```

```python
plt.subplot(211)
plt.hist(asia_data_2007.gdpPerCapita, range=(0,50000) ,edgecolor= "black")
plt.ylabel("Asia")
plt.subplot(212)
plt.ylabel("Europe")
plt.hist(Europe_data_2007.gdpPerCapita, range=(0,50000),edgecolor= "red")
plt.show()
```

### Histograms show that Europe have more countries with high and medium GDp, unlike Asia where most countries with low GDp



# 2nd problem: compare Europe and Americas life expectency

```python

americas=data[data.continent=="Americas"]
Europe=data[data.continent=="Europe"]
americas_1997=americas[americas.year== 1997]
Europe_1997=Europe[Europe.year== 1997]
```

```python
bins=20
plt.subplot(211)
plt.hist(americas_1997.lifeExpectancy , bins, range=(55,85), edgecolor="black")
plt.ylabel("Americas")
plt.subplot(212)
plt.hist(Europe_1997.lifeExpectancy , bins, range=(55,85), edgecolor="black")
plt.ylabel("Europe")
plt.show()
```
